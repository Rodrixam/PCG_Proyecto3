```python
import os
import fileinput
import random

folder_path = os.path.join('/content', 'TheVGLC', 'Super Mario Bros', 'Processed')
folder_contents = os.listdir(folder_path)
```

```python
def peek_line(n):
  pos = n.tell()
  line = n.readline()
  n.seek(pos)
  return line

X = []
Y = []

column_height = 14

for file_name in folder_contents:
  file_path = os.path.join(folder_path, file_name)
  with open(file_path, 'r') as file:
    sequence_length = len(peek_line(file))
    level_text = file.read()

    for i in range(0, sequence_length):
      for j in range(0, column_height):
        if i != 0:
          x1 = level_text[(i-1)+j*sequence_length]
        else:
          x1 = 'N'
        if j != column_height - 1:
          x2 = level_text[i + (j+1) * sequence_length]
        else:
          x2 = 'N'

        y = level_text[i + j * sequence_length]
        X.append([x1, x2])
        Y.append(y)
#X
#Y
```

```python
all_characters_set = set()
chars_value = dict()
value_chars = dict()

for tup in X:
    all_characters_set.add(tup[0])
    all_characters_set.add(tup[1])


integer_mapping = {x: i for i,x in enumerate(all_characters_set)}

vec = [integer_mapping[word] for word in all_characters_set]
all_characters_set = list(all_characters_set)
for i in vec:
  chars_value[all_characters_set[i]] = i
  value_chars[i] = all_characters_set[i]

all_characters_set
```
